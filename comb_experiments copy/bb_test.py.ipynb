```python
#random graph tests
import numpy as np
from pgmpy.models import MarkovNetwork
from pgmpy.inference import Mplp
from pgmpy.factors.discrete import DiscreteFactor
student = MarkovNetwork()

N = 5
p = 0.2
k = 2
mode = 'cut'
for i in range(N):  
    student.add_node(str(i))
    if mode == 'random': 
        node_factor = DiscreteFactor([str(i)], cardinality=[2], values=np.random.uniform(0.0,1.0,k))
    elif mode == 'cut': 
        node_factor = DiscreteFactor([str(i)], cardinality=[2], values=np.random.uniform(0.0,1.0,k))
        #node_factor = DiscreteFactor([str(i)], cardinality=[2], values=np.zeros(k))
    student.add_factors(node_factor)
    
for i in range(N):
    for j in range(i+1,N):
        ind = np.random.binomial(1,p)
        if ind == 1:
            student.add_edge(str(i),str(j))
            if mode == 'random': 
                edge_factor = DiscreteFactor([str(i), str(j)], cardinality=[2, 2],values=np.random.uniform(0.0,1.0,k**2))
            elif mode == 'cut': 
                edge_factor = DiscreteFactor([str(i), str(j)], cardinality=[2, 2],values=[0,1,1,0])
            student.add_factors(edge_factor)
            
mplp = Mplp(student)
import sys
sys.path.append('/opt/anaconda3/pkgs/pgmpy-0.1.20-py37_0/lib/python3.7/site-packages/pgmpy/inference')
from branch_bound import BB
a = BB(mplp)
a.run()
```
